# ***REGRESSION***

```python
import pandas as pd
import statsmodels.formula.api as sm

# Load your dataset
air = pd.read_csv("https://raw.githubusercontent.com/cmvigneswaranaero-gif/AC_PERFROMANCE_WITH_ML/main/New_Atm_Para.csv")

# 1. Model 1: Altitude vs Density

model1 = sm.ols(formula='Density ~ Altitude', data=air)
fitted1 = model1.fit()
print("\nRegression Summary: Altitude vs Density")
print(fitted1.summary())


```

## ***R Squared***

```python
import pandas as pd
import statsmodels.formula.api as sm

# Load dataset
air = pd.read_csv("https://raw.githubusercontent.com/cmvigneswaranaero-gif/AC_PERFROMANCE_WITH_ML/main/New_Atm_Para.csv")

# 1. Model 1: Altitude vs Density

model1 = sm.ols(formula='Density ~ Altitude', data=air)
fitted1 = model1.fit()
print("\nRegression Summary: Altitude vs Density")
print(fitted1.summary())


# 2. Model 2: Density vs Temperature

model2 = sm.ols(formula='Density ~ Temperature', data=air)
fitted2 = model2.fit()
print("\nRegression Summary: Density vs Temperature")
print(fitted2.summary())


# 3. Model 3: Density vs Pressure

model3 = sm.ols(formula='Density ~ Pressure', data=air)
fitted3 = model3.fit()
print("\nRegression Summary: Density vs Pressure")
print(fitted3.summary())


# 4. Model 4: Density vs Wind_Gust

model4 = sm.ols(formula='Density ~ Wind_Gust', data=air)
fitted4 = model4.fit()
print("\nRegression Summary: Density vs Wind_Gust")
print(fitted4.summary())


# prediction through diffrent model
# below value taken from ISA

# Altitude      =  8900
# Temperature   =   230.3
# Pressure      = 31195.25
# Wind_Gust     = 11.43
# Density       = 0.47197

# Predict Density for Altitude = 8900 m
new_data1 = pd.DataFrame({'Altitude': [8900]})
pred1 = fitted1.predict(new_data1)
print(f"\nPredicted Density at 8900 m Altitude: {pred1[0]:.4f} kg/m³")

# Predict Density for Temperature = 230.3 K
new_data2 = pd.DataFrame({'Temperature': [230.3]})
pred2 = fitted2.predict(new_data2)
print(f"\nPredicted Density at Temperature 230.3 K: {pred2[0]:.4f} kg/m³")

# Predict Density for Pressure = 31195.25 Pa
new_data3 = pd.DataFrame({'Pressure': [31195.25]})
pred3 = fitted3.predict(new_data3)
print(f"\nPredicted Density at Pressure 31195.25 Pa: {pred3[0]:.4f} kg/m³")

# Predict Density for Wind_Gust = 11.43 m/s
new_data4 = pd.DataFrame({'Wind_Gust': [11.43]})
pred4 = fitted4.predict(new_data4)
print(f"\nPredicted Density at Wind Gust 11.43 m/s: {pred4[0]:.4f} kg/m³")

```

# ***VIF***

```python
#Code for VIF Calculation
#Writing a function to calculate the VIF values
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# -------------------------------------------------------------
# Load and clean dataset
# -------------------------------------------------------------
data = pd.read_csv("https://raw.githubusercontent.com/cmvigneswaranaero-gif/AC_PERFROMANCE_WITH_ML/refs/heads/main/VIF.csv")
data.columns = data.columns.str.strip()  # remove extra spaces

# Select input features (independent variables)
X_Data = data[['Combustion_temp', 'Exhaust_velocity', 'percentage_of_mdota', 'percentage_of_mdotf']]

# -------------------------------------------------------------
# Function to calculate VIF
# -------------------------------------------------------------
def vif_cal(input_data):
    X_with_const = sm.add_constant(input_data)   # add intercept
    vif_data = pd.DataFrame()
    vif_data["Feature"] = X_with_const.columns
    vif_data["VIF"] = [variance_inflation_factor(X_with_const.values, i)
                       for i in range(X_with_const.shape[1])]
    return vif_data

# -------------------------------------------------------------
# Calculate VIF for all variables
# -------------------------------------------------------------
print("VIF for all variables:")
print(vif_cal(input_data=X_Data))


```

```python
print("\nVIF after dropping 'Combustion_temp':")
print(vif_cal(input_data=X_Data.drop(["Combustion_temp"], axis=1)))
```

```python
print("\n VIF after dropping 'Combustion_temp' and 'percentage_of_mdotf':")
print(vif_cal(input_data=X_Data.drop(['Combustion_temp', 'percentage_of_mdotf'], axis=1)))
```

```python
# regression model dropping combustion_temp & percentage_of_mdotf
model = smf.ols(formula='Avg_thrust ~ Exhaust_velocity+ percentage_of_mdota', data=data)

# Fit the model
fitted2 = model.fit()

# Display the regression summary
print(fitted2.summary())
```
